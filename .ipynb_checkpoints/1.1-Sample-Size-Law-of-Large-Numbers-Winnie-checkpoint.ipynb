#### 1. How does this prove the Law of Large Numbers?

As mentioned in the definition, the law of large numbers states that the posteriori probability grows closer to the priori probability with the increase in the number of trials. 

As shown in the codebit under "Sampling the Dataset and Analysis", the probability tends to reach a more stable result, going from 100% (@n=1) down to 60%(@n=10) down to ~50%(@n>1000). The possibility of the same event reoccuring due to "luck" (e.g. getting 2 heads out of 2 coin flip trials) is greatly diminished and is balanced out by numerous randomized events of the said scenario. By increasing the number of trials, we are slowly able to get more information, which gurantees a "more consistent" result. 

#### 2. Does the law of large numbers apply for all datasets?

The law of large numbers apply for all datasets assuming that there is sufficient data that will be used as a basis for predicting the probabilities. By steadily increasing the number of samples, the actual probability will always produce a more stable result, effectively producing the expected probability.

#### 3. How else can we use the law of large numbers?

The law of large numbers can be applied in other scenario that require are random (or requires "luck") such as rolling a certain value in a dice, or getting a certain card in a shuffled deck. In a deeper sense, the law of large numbers may also be used in predicting probability of losses or gains from a sale based from historical data (assuming that there is sufficient data to be used as basis).

## Coinflips Sampling

When flipping a fair coin, one can expect it to land on 'heads' with a probability of 50%. We want to show the Law of Large Numbers here by flipping the coin a large number of times and see how the probability moves towards 50%.

```python
# TODO:
# Generate n random coinflips using python using the np.random.randint function with probability 50%
# Example: the list "data" should contain n integers that are either 0 or 1
# Find the a posteriori probability for the coinflips.
def coinflip(n):
    data = []

    for i in range(1,n):
        a = np.random.randint(0,2)   
        data.append(a)
    return data

n = np.random.randint(1, 1000000)
data = coinflip(n)
hCount = data.count(HEADS)
probability = hCount/n * 100
print("Posteriori probability for coinflip using", n, "trials:", round(probability, 5), "%")
```

## Function Creation

To simplify the workflow and process of the experiments, we will create functions for random sampling from our dataset and finding the a posteriori probability from our sample.

```python
# TODO:
# Create a function that randomly takes sampleSize items of the list data and places them into the list "samples"
# Use the np.random.choice function.

def getSample(sampleSize, data):
    samples = np.random.choice(data, sampleSize)
    #print(samples)
    samples = np.ndarray.tolist(samples)
    return samples

# Create a function that finds the a posteriori probability for the coinflips samples.

def aPosteriori(sample):
    n = len(sample)
    hCount = sample.count(HEADS)
    probability = hCount/n * 100
    return probability
```

## Sampling the Dataset and Analysis

Now that the data has been generated and the functions are created, the dataset only needs to be sampled and we can analyze our found information from the data.

```python
# TODO:
# Prove the Law of Large Numbers.
# Show that the a posteriori probability converges to the a priori probability as the number of trials increases.

sampleSize = 1
n = 1000000
data = coinflip(n)

print("N = ", n)
print("aPriori for a fair coin flip is 50%\n")
# YOUR CODE HERE
while sampleSize <= n:
    sampleData = getSample(sampleSize, data)
    #print(sampleData)
    print("At n=", sampleSize)
    print("\taPosteriori Probability of getting HEADS is", round(aPosteriori(sampleData), 5), "%")
    sampleSize*=10
```

## Sampling Means and Population Means

The law of large numbers also works for samples from a probability distribution. By applying the law of large numbers to sampling a population, one can see that the sample mean converges to the distribution mean as the sample size increases.

```python
#generate n random values from a beta distribution
n = 100000
a = 5
b = 1
data = np.random.beta(a,b,n)

#visualize the histogram of the data
fig = plt.figure(figsize=(15,5))
num, bins, patches = plt.hist(data, 20,normed=True)

mean = np.mean(data);
stdev = np.std(data);

#plot the corresponding to the probability density function
density = gaussian_kde(data)
xs = np.linspace(np.min(data),np.max(data),200)
plt.plot(xs, density(xs), 'r--',linewidth=2)

#plot labels
plt.xlabel('Values')
plt.ylabel('Density')
plt.title(r'$\mathrm{Histogram:}\ \mu=%.3f,\ \sigma=%.3f$' %(mean, stdev))
plt.grid(True)


plt.show()
```
